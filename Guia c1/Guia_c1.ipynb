```python
# P1 c1 otoño 2023

"""
Se dispone de una funcion sgte_dato() que cada vez que se llama
entrega el siguiente numero de una secuencia (no nulo). Esa se-
cuencia termina cuando cuando sgte_dato() entrega un 0.

hay que escribir un programa que reciba todo esos datos y que 
imprima todos los numeros positivos en orden de aparición y 
los numeros negativos en orden inverso.

solo se puede usar de estructuras de datos las pilas y las colas 
"""

class NodoLista:
    def __init__(self,info,sgte=None):
        self.info=info
        self.sgte=sgte
class Pila:
    def __init__(self):
        self.tope = None
    def push(self,x):
        self.tope = NodoLista(x, self.tope)
        
    def pop(self):
        assert self.tope is not None
        
        x = self.tope.info
        self.tope = self.tope.sgte
        
        return x
    
    def is_empty(self):
        return self.tope is None

class Cola:
    def __init__(self):
        self.primero=None
        self.ultimo=None
    def enq(self,x):
        p=NodoLista(x)
        if self.ultimo is not None: # cola no vacía, agregamos al final
            self.ultimo.sgte=p
            self.ultimo=p
        else: # la cola estaba vacía
            self.primero=p
            self.ultimo=p
    def deq(self):
        assert self.primero is not None
        x=self.primero.info
        if self.primero is not self.ultimo: # hay más de 1 elemento
            self.primero=self.primero.sgte
        else: # hay solo 1 elemento, el deq deja la cola vacía
            self.primero=None
            self.ultimo=None
        return x
    def is_empty(self):
        return self.primero is None
    
```

```python
from random import randint as rndm

def sgte_dato():
    return rndm(-6,6)


dato = sgte_dato()
print(f'{dato} fuera de while')
pos = Cola()
neg = Pila()
while dato != 0:
    
    if dato > 0:
        pos.enq(dato)
    
    if dato < 0:
        neg.push(dato)
    dato = sgte_dato()
    print(f'{dato}')
    
while not neg.is_empty():
    print(neg.pop())
    
while not pos.is_empty():
    print(pos.deq())
```
